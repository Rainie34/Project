    - US dollars per capita
    - Type of financing as % of total health spending
- Year
- Flag code: An observation on the data
    - B: Break
    - D: Difference in methodology
    - E: Estimated value
    - P: Provisional value
       
For the purposes of our study, we will be using the percentage of GDP as a measure as US dollars per capita does not consider differences in purchasing power between countries, which might inflate the values of healthcare expenditure for more developed countries.

# 3 Data Loading and Selection

First, we start by loading the data from the 4 different sources we are working with. We have dataframes for health expenditure, life expectancy, country codes, and regions. Also, we will load the necessary packages for the study.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy.stats import linregress

# Health expenditure data
he_raw = pd.read_csv('https://raw.githubusercontent.com/Juanfra21/DAV-5400/main/health_expenditure.csv')

# Life expectancy data
le_raw = pd.read_csv('https://raw.githubusercontent.com/Juanfra21/DAV-5400/main/life_expectancy.csv')

# Read the wikipedia page with the country codes
country_codes_raw = pd.read_html('https://en.wikipedia.org/w/index.php?title=List_of_ISO_3166_country_codes&oldid=1180314150',header=1)

# Read the wikipedia page with the regions
regions_raw = pd.read_html('https://en.wikipedia.org/w/index.php?title=List_of_countries_by_the_United_Nations_geoscheme&oldid=1165741636')
```

## 3.1 Data Loading

### 3.1.1 Health Expenditure Data

For the health expenditure data, we will select only the necessary columns, then transform it from a long format into a wide format, and finally set new column names for easier visualization and analysis.

```python
# Select only the necessary columns
he_raw = he_raw.drop(columns=[
    'INDICATOR',
    'FREQUENCY',
    "Flag Codes"])

# We use pivot table to create a table in a wide format
he_raw = he_raw.pivot_table(index=["LOCATION", "TIME"],
                    columns=["MEASURE", "SUBJECT"],
                    values="Value",
                    aggfunc="sum",
                    fill_value=np.nan) # We fill the missing values with NaN

# Reset indexes
he_raw = he_raw.reset_index()

# Set new column names
he_new_column_names = ['country',
                    'year',
                    'compulsory_gdp',
                    'oop_gdp',
                    'total_gdp',
                    'voluntary_gdp',
                    'compulsory_percent',
                    'oop_percent',
                    'voluntary_percent',
                    'compulsory_usd',
                    'oop_usd',
                    'total_usd',
                    'voluntary_usd']

he_raw.columns = he_new_column_names

he = he_raw[['country',
         'year',
         'compulsory_gdp',
         'oop_gdp',
         'total_gdp',
         'voluntary_gdp']]

# Show the results
he.head()
```

### 3.1.2 Life Expectancy Data

For life expectancy data, we will select only the necessary columns and give them new names for easier visualization and analysis.

```python
le = le_raw.loc[(le_raw["SUBJECT"] == "TOT")]

# Select only the necessary columns
le = le.drop(columns=[
    "INDICATOR",
    "FREQUENCY",
    "MEASURE",
    "SUBJECT",
    "Flag Codes"])

# Set new column names
le_new_column_names = ["country",
                       "year",
                    "life_expectancy"]

le.columns = le_new_column_names

# Reset indexes
le = le.reset_index(drop = True)

# Show the results
le.head()
```

### 3.1.3 Country Code Data

For country codes, we will first select the first table on the Wikipedia page, then delete all the reference numbers within brackets using regex, and finally select the necessary columns and rename them for easier work and visualization.

```python
# Retrieve the first table in the page
country_codes = country_codes_raw[0]

# Use regex to delete reference numbers which are between brackets
country_codes = country_codes.replace(r'\[.*?\]', '', regex=True)

# Select only the necessary columns
country_codes = country_codes.drop(columns=[
    'Official state name[6][a]',
    'Sovereignty[6][7][8]',
    'Alpha-2 code[5]',
    'Numeric code[5]',
    'Subdivision code links[3]',
    'Internet ccTLD[9]'])

# Standardize column names
country_codes = country_codes.rename(columns={
    'Country name[5]': 'country_name',
    'Alpha-3 code[5]': 'country_code'})

# Show the results
country_codes.head()
```

### 3.1.4 Regions Data

For regions data, we will select the first table on the page, delete all the reference numbers using regex, and select the necessary columns, renaming them according to the convention used before.

```python
# Retrieve the first table in the page
regions = regions_raw[0]

# Use regex to delete reference numbers which are between bracket
regions = regions.replace(r'\[.*?\]','', regex=True)

# Select only the necessary columns
regions = regions.drop(columns=['UNSD M49 Codes','Sub-region','Intermediate Region'])

# Standardize column names
regions = regions.rename(columns={
    'Country or Area': 'country_name',
    "Region":"region"})

# Show the results
regions.head()
```

## 3.2 Year Selection

### 3.2.1 Health Expenditure

We start the analysis for year selection by checking how many null values are there in the health expenditure data by year, and also the number of unique countries per year in the health expenditure dataframe.

```python
# Let's count the number of NaNs values per year accross all columns
he_nan_count_by_column = he.groupby("year").apply(lambda x: x.isna().sum().sum()).reset_index(name="Number of NaNs")

# Set Seaborn style
sns.set(style="whitegrid")

# Plotting with Seaborn
plt.figure(figsize=(10, 4))
sns.barplot(x="year", y="Number of NaNs", data=he_nan_count_by_column, palette="viridis")

# Adding labels and title
plt.xlabel("Year")
plt.ylabel("NaN Count")
plt.title("Total NaN Count by Year")

# Rotate x-axis labels
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

```python
# We select the number of unique countries per year in the health expenditure data frame
he_unique_country_by_year = he.groupby('year')['country'].nunique().reset_index()

# Set Seaborn style
sns.set(style="whitegrid")

# Plotting with Seaborn
plt.figure(figsize=(10, 4))
sns.barplot(x='year', y='country', data=he_unique_country_by_year, palette="viridis")

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Number of Unique Countries')
plt.title('Number of Unique Countries per Year in the Health Expenditure Data Frame')

# Rotate x-axis labels
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

As it can be seen, the years from 2011 to 2018 have 0 null values, whereas the years from 2003 to 2019 have the highest number of unique countries, with up to 50 unique countries in these years.

### 3.2.2 Life Expectancy

```python
# Let's count the number of NaNs values per year accross all columns
le_nan_count_by_column = le.groupby("year").apply(lambda x: x.isna().sum().sum()).reset_index(name="Number of NaNs")

# Set Seaborn style
sns.set(style="whitegrid")

# Plotting with Seaborn
plt.figure(figsize=(10, 4))
sns.barplot(x="year", y="Number of NaNs", data=le_nan_count_by_column, palette="viridis")

# Adding labels and title
plt.xlabel("Year")
plt.ylabel("NaN Count")
plt.title("Total NaN Count by Year")

# Rotate x-axis labels
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

```python
# We select the number of unique countries per year in the life expectancy data frame
le_unique_country_by_year = le.groupby('year')['country'].nunique().reset_index()

# Set Seaborn style (optional)
sns.set(style="whitegrid")

# Plotting with Seaborn
plt.figure(figsize=(10, 4))
sns.barplot(x='year', y='country', data=le_unique_country_by_year, palette="viridis")

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Number of Unique Countries')
plt.title('Number of Unique Countries per Year in the Life Expectancy Data Frame')

# Rotate x-axis labels
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

When it comes to life expectancy, we can see that this dataframe does not have null values. Additionally, the largest number of unique countries occurs between 2002 up to 2019, with 48 unique countries in these years.

Considering this, we decided to select the year 2019 as the most recent year with a low number of null values and a high number of unique countries.

### 3.2.3 Selected Year Data Frame Selection

Now, using the loc function, we select the data from the health expenditure dataframe and life expectancy dataframe for the year 2019.

```python
# Select data from the year 2019
he_2019 = he.loc[he["year"] == 2019].reset_index(drop = True)

# We drop the year column from this data frame
he_2019.drop("year", axis=1, inplace=True)

# Show the results
he_2019.head()
```

```python
# Select data from the year 2019
le_2019 = le.loc[le["year"] == 2019].reset_index(drop = True)

# We drop the year column from this data frame
le_2019.drop("year", axis=1, inplace=True)

# Show the results
le_2019.head()
```

## 3.3 Merging

### 3.3.1 Health Expenditure and Life Expectancy

Having selected the data from 2019, we merge the dataframes into the health expenditure dataframe, starting with the life expectancy data.

```python
# We first merge the health expenditure and life expectancy data from the year 2019
df = pd.merge(he_2019, le_2019, on=['country'], how = 'left')

# Show the results
df.head()
```

### 3.3.2 Country Codes

We follow up by merging the country codes into the previously resulting dataframe.

```python
# We merge the health expenditure and life expectancy data from the year 2019 with the country codes
df = pd.merge(df, country_codes, left_on=['country'], right_on = ['country_code'], how = 'left')

# We drop the country codes
df.drop('country_code', axis=1, inplace=True)

# Show the results
df.head()
```

### 3.3.3 Regions

And we finish up by merging the regions into the previously resulting dataframe.

```python
# We remove all starting and ending spaces from country name column in both data frames to make sure this doesn't interfere with the merge
df["country_name"] = df["country_name"].str.strip()
regions["country_name"] = regions["country_name"].str.strip()

# We merge the data frame with the regions on country name
df = pd.merge(df, regions, on=['country_name'], how = 'left')

# Show the results
df.head()
```

# 7 Investigative Analysis & Results

To deepen the analysis made in the EDA, with scipy.stats package, we will calculate the Pearson correlation coefficient between our attributes and life expectancy to check if there is, in fact, any **linear** correlation or not.

The value of the Pearson correlation coefficient indicates whether the correlation is positive or negative between the variables and also whether it is strong or weak, according to the values in this table:

| Correlation coefficient | Correlation strength | Correlation type |
|-------------------------|----------------------|------------------|
| -.7 to -1               | Very strong          | Negative         |
| -.5 to -.7              | Strong               | Negative         |
| -.3 to -.5              | Moderate             | Negative         |
| 0 to -.3                | Weak                 | Negative         |
| 0                       | None                 | Zero             |
| 0 to .3                 | Weak                 | Positive         |
| .3 to .5                | Moderate             | Positive         |
| .5 to .7                | Strong               | Positive         |
| .7 to 1                 | Very strong          | Positive         |

If a strong correlation is observed, we will perform a linear regression to assess how much life expectancy varies based on changes in one of the compared variables.

When evaluating statistical significance using the p-value, we consider a threshold of 0.05 or lower. This convention stems from the standard practice in hypothesis testing, where a p-value below 0.05 indicates that the observed results are unlikely to have occurred by random chance alone. In other words, a p-value of 0.05 or lower suggests that there is sufficient evidence to reject the null hypothesis in favor of the alternative hypothesis. This significance level of 0.05 is widely accepted in scientific research as a reasonable balance between avoiding type I errors (incorrectly rejecting a true null hypothesis) and type II errors (failing to reject a false null hypothesis), providing a common and practical criterion for interpreting the results of statistical tests.
